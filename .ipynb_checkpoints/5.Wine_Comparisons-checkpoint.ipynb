## Multiple Group Counts

```python
from ipywidgets import interact, fixed
```

```python
def pivot_count(df, rows, columns):
    df_pivot = df.pivot_table(values="pH", # could be any column, since we are just counting rows 
                              index=rows, 
                              columns=columns, 
                              aggfunc=np.size
                             ).dropna(axis=0, how='all')
    return df_pivot
```

```python
quality_by_alcohol = pivot_count(wine, "alcohol_level","quality_rating")
quality_by_alcohol
```

```python
quality_by_pH = pivot_count(wine, "pH_level","quality_rating")
quality_by_pH
```

```python
def heatmap_count_types(df, col1, col2):
    pivot_table = pivot_count(df,col1, col2)
    sns.heatmap(pivot_table, annot=True, fmt='g')
    plt.ylabel(col1)
    plt.xlabel(col2)
    plt.show()
```

```python
heatmap_count_types(wine, "alcohol_level","density_level")
```

```python
heatmap_count_types(wine, "alcohol_level","chloride_level")
```

```python
heatmap_count_types(wine, "pH_level","citric_acid_level")
```

## Categorical Variable Means
Next we will calculate the means of the categorical groups of numerical variables after removing outliers.

```python
def means_by_category(col_group, col_calc):
    wine_no_outliers = wine[wine[col_calc]<wine[col_calc].quantile(0.99)]
    wine_no_outliers.groupby(col_group)[col_calc].mean().plot.barh()   
    plt.ylabel(col_group)
    plt.xlabel('Mean values for variable {}'.format(col_calc))
    plt.show()
```

```python
numerical_variables = wine.select_dtypes(include=[np.number]).columns
categorical_variables = wine.select_dtypes(include=['object', 'category']).columns
columns_type = [col for col in wine.columns if col.endswith("_type")]
```

```python
interact(means_by_category, col_group=categorical_variables, col_calc=numerical_variables);
```

#### Conclusions

- High volatile acidity are general rated as 'bad' quality
- Wines with high levels of citric acid and chlorides are generally rated as 'excellent' quality
- Low pH = High citric acid levels, as noted in notebook 3
- Wines with higher volatile acidity, sulfur dioxide, and chlorides have a lower alcohol content. Those with higher citric acid levels and sulfates have a higher alcohol content. Wines with a higher alcohol content are rated higher for quality.
- Wines with a very high density have high levels of sulfates, chlorides, sulfur dioxide, and sugar, but have a lower alcohol content
- Wines with very high sugar levels have higher levels of citric acid.

### Means by multiple groups

```python
def pivot_mean(rows, columns, calc_field):
    df_pivot = wine.pivot_table(values=calc_field, 
                              index=rows, 
                              columns=columns, 
                              aggfunc=np.mean
                             ).dropna(axis=0, how='all')
    return df_pivot
```

```python
pivot_mean("alcohol_level","quality_rating", "pH")
```

```python
def heatmap_means_types(col1, col2, col3):
    pivot_table = pivot_mean(col1, col2, col3)
    sns.heatmap(pivot_table, annot=True, fmt='g')
    plt.ylabel(col1)
    plt.xlabel(col2)
    plt.show()
```

```python
interact(heatmap_means_types, col1=wine.columns, col2=wine.columns, col3=numerical_variables);
```

#### Conclusions
* Wines with low alcohol levels and high pH levels have higher volatile acidity and higher levels of chlorides.
* Wines with very high alcohol_levels and low pH levels have higher citric acid and sulfur dioxides levels and lower density.
* **Wines with lower pH and higher alcohol content are rated as a higher quality.**

## Trendline Comparison

Normally a trendline comparison is done in a data set which has a temporal variable. In this data set we will replace the "time" component with "quality" to understand how the varying components of a wine change in wines in increasing quality.

```python
wine.shape
```

```python
wine.head()
```

```python
wine.groupby('quality')['alcohol'].mean().plot();
```

```python
def mean_trend(col_calc):
    wine.groupby('quality')[col_calc].mean().plot()
    plt.ylabel(col_calc)
    plt.title("{} level with increase in quality".format(col_calc))
    plt.show()
```

```python
interact(mean_trend, col_calc=numerical_variables);
```

```python
def count_trend(col_calc):
    for category in wine[col_calc].unique():
        wine_alcohol = wine[wine[col_calc]==category].groupby(
            'residual_sugar').apply(np.size)
        try:
            plt.plot(
                wine_alcohol.index,
                wine_alcohol,
                label=category
            )
        except ZeroDivisionError:
            pass
        plt.legend()
```

```python
count_trend("quality_rating")
```

```python
count_trend("pH_level")
```

### Conclusions
* As quality of a wine increases, chlorides, pH, density, and volatile acidity levels decrease.
* As quality of a wine increases, sulfate and citric acid levels increase.
* As alcohol content increases, quality rating does too.

## Analysis Conclusion
### Description
This data set contains the levels of various components of wine as well as the quality on a scale of 0 to 10.
The original data set is located: https://www.kaggle.com/piyushgoyal443/red-wine-dataset
The entities associated with the data set are the id
### Data Dictionary
The variables that exist in the data set are:
* volatile_acidity (float64): the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste
* citric_acid (float64): found in small quantities, citric acid can add 'freshness' and flavor to wines
* residual_sugar (float64): the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet
* chlorides (float64): the amount of salt in the wine
* total_sulfur_dioxide (float64): amount of free and bound forms of SO2; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine
* density (float64): the density of water is close to that of water depending on the percent alcohol and sugar content
* pH (float64): describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale
* sulfates (float64): a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant
* alcohol (float64): the percent alcohol content of the wine
* quality (int64): score between 0 and 10
* quality_rating (category)
* quality_rating (category)
* pH_level (category)
* alcohol_level (category)
* density_level (category)
* citric_acid_level (category)
* sugar_level (category)
* chloride_level (category)
* total_sulfur_level (category)

### QA

* There are no duplicate rows or missing values
* The majority of wines are rated at a 5 or 6 on a 0 to 10 scale
* There are outliers in most variables, specifically chlorides, sulfur/sulfates, and residual sugars.

### Distribution of Variables
* None of the variables follow a normal distribution.
* The variable `quality` has an unbalanced discrete distribution (the values 5, 6 and 7 account for 95% of the wines). It has been grouped into 'bad', 'okay', 'good', and 'excellent'.
* The variable 'citric acid' may also have an unbalanced distribution. The value 0.0 accounts for 8% of the wine's citric acid content. This sounds like a small amount, but there are lots of values.
* 28% of wines have very low alcohol content
* 23% of wines have very low fixed acidity
* 1/4 of wines have low sugar levels

### Comparisons

* There is an inverse relationship between alcohol_level and density_level as well as with chloride_level
* There is an inverse relationship between pH_level and citric_acid_level
* Wines with low alcohol levels and high pH levels have higher volatile acidity and higher levels of chlorides.
* Wines with very high alcohol_levels and low pH levels have higher citric acid and sulfur dioxides levels and lower density.
* **Wines with lower pH and higher alcohol content are rated as a higher quality.**
* As quality of a wine increases, chlorides, pH, density, and volatile acidity levels decrease.
* As quality of a wine increases, sulfate and citric acid levels increase.
* As alcohol content increases, quality rating does too.
* None of the variables have a high correlation. pH and citric acid have an inverse correlation (-0.54) but this was noted in the variable grouping analysis.
